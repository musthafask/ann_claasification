#Problem Statment

The problem statment of this dataset is predicting that the insurance claim by the customer is reported fraud or not

```python
import pandas as pd
import numpy as np
```

# LOADING THE DATASETS

```python
# Load the first dataset
df1 = pd.read_csv('/content/Train_Claim.csv')
```

```python
df1.shape
```

```python
# Load the second dataset
df2 = pd.read_csv('/content/Train_Demographics.csv')
```

```python
df2.shape
```

```python
# Load the third dataset
df3 = pd.read_csv('/content/Train_Policy.csv')
```

```python
# Load the fourth dataset
df4 = pd.read_csv('/content/Train_Vehicle.csv')
```

```python
df4.shape
```

```python

df4=df4.pivot(index='CustomerID',columns='VehicleAttribute',values='VehicleAttributeDetails').reset_index()
```

```python
df4
```

```python
# Load the fifth dataset
df5 = pd.read_csv('/content/Traindata_with_Target.csv')
```

# MERGING THE DATA

```python
# Merge the datasets using a common column
merged_df = pd.merge(df1, df2, on='CustomerID', how='inner')
merged_df = pd.merge(merged_df, df3, on='CustomerID', how='inner')
merged_df = pd.merge(merged_df, df4, on='CustomerID', how='inner')
merged_df = pd.merge(merged_df, df5, on='CustomerID', how='inner')

# Print the merged dataset
print(merged_df)

# Save the merged dataset to a new file
#merged_df.to_csv('merged_dataset.csv', index=False)
```

```python
merged_df
```

#INORMATION OF DATA

```python
merged_df.info()
```

```python
merged_df.shape
```

# CHECKING FOR NULL VALUES

```python
for i in merged_df:
  print('\n',i,merged_df[i].unique())
```

```python
merged_df['PoliceReport'] = merged_df['PoliceReport'].replace('?', np.nan)
merged_df['PropertyDamage'] = merged_df['PropertyDamage'].replace('?', np.nan)
merged_df['TypeOfCollission'] = merged_df['TypeOfCollission'].replace('?', np.nan)
merged_df['InsuredGender'] = merged_df['InsuredGender'].replace('Cust10000', np.nan)
merged_df['Country'] = merged_df['Country'].replace('Cust10000', np.nan)
```

```python
merged_df['PoliceReport']
```

```python
merged_df['TypeOfCollission']
```

```python
merged_df.isnull().sum()
```

# FILLING THE NULL VALUES

```python
for cols in merged_df:
  if merged_df[cols].dtypes=='object':
    merged_df = merged_df.fillna(merged_df[cols].value_counts().index[0])
```

```python
merged_df.isnull().sum()
```

```python
merged_df.head()
```

# DUMMIFICATION OF CATEGORICAL VARIABLES

```python
cols_to_dummify = ['TypeOfIncident', 'PoliceReport', 'PropertyDamage', 'InsurancePolicyState','IncidentState', 'IncidentCity', 'AuthoritiesContacted', 'SeverityOfIncident', 'InsuredRelationship', 'BodilyInjuries', 'InsuredGender', 'TypeOfCollission' ]
merged_df = pd.get_dummies(merged_df, columns=cols_to_dummify, prefix=cols_to_dummify)
```

```python
merged_df.head()
```

```python
merged_df['ReportedFraud'] = merged_df['ReportedFraud'].str.replace('Y', '1', n=-1)
merged_df['ReportedFraud'] = merged_df['ReportedFraud'].str.replace('N', '0', n=-1)
```

# DROPPING THE COULMNS 

```python
d_col = ['InsuredOccupation', 'InsuredHobbies', 'IncidentAddress', 'IncidentTime', 'InsuredZipCode', 'InsuredEducationLevel', 'VehicleAttributeDetails', 'VehicleAttribute', 'NumberOfVehicles', 'Policy_CombinedSingleLimit', 'Country', 'Witnesses']
merged_df_new = merged_df.drop(d_col , axis=1).copy()
```

```python
merged_df_new.columns
```

```python
merged_df_new.head()
```

# TEXT TO VALUE

```python
merged_df_new['CustomerID'] =merged_df_new['CustomerID'].str.replace('Cust', '').astype(int)
```

```python
merged_df_new.head()
```

# COVERTING 

```python
merged_df_new['DateOfIncident'] = pd.to_datetime(merged_df_new['DateOfIncident'])
```

```python
merged_df_new['year_of_incedence'] = merged_df_new['DateOfIncident'].dt.year
merged_df_new['month_of_incedence'] = merged_df_new['DateOfIncident'].dt.month
merged_df_new['day_of_incedence'] = merged_df_new['DateOfIncident'].dt.day
merged_df_new.drop('DateOfIncident', axis=1, inplace=True)
```

```python
merged_df_new.head()
```

```python
merged_df_new['DateOfPolicyCoverage'] = pd.to_datetime(merged_df_new['DateOfPolicyCoverage'])
merged_df_new['year_of_policycoverage'] = merged_df_new['DateOfPolicyCoverage'].dt.year
merged_df_new['month_of_policycoverage'] = merged_df_new['DateOfPolicyCoverage'].dt.month
merged_df_new['day_of_policycoverage'] = merged_df_new['DateOfPolicyCoverage'].dt.day
merged_df_new.drop('DateOfPolicyCoverage', axis=1, inplace=True)
```

```python
merged_df_new.head()
```

```python
# Count the number of duplicated rows
duplicated_values = merged_df_new.duplicated().sum()
print("Number of duplicated rows:", duplicated_values)

# Drop the duplicated rows and reset the index
merged_df_new.drop_duplicates(inplace=True)
merged_df_new.reset_index(drop=True, inplace=True)

# Check the number of rows after dropping duplicates
print("Number of rows after dropping duplicates:", len(merged_df_new))
```

# CHANGING TYPES

```python
merged_df_new['ReportedFraud'] = merged_df_new['ReportedFraud'].astype(int)
merged_df_new['AmountOfTotalClaim'] = pd.to_numeric(merged_df_new['AmountOfTotalClaim'], errors='coerce').astype('Int64')

```
